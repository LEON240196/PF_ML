```python
import gradio as gr
from joblib import load
import numpy as np
model = load('model_svm.joblib')

def run_my_model(gender, age, height, weight, name, family_history):
    x = np.array([gender, age, height, weight,family_history, 
    'no', 2,3, 'Sometimes', 'no', 2, 'no', 0, 1, 
                'no', 'Public_Transportation'])
    x = x.reshape(1,-1)
    x = pd.DataFrame(x, columns=X.columns)
    y_pred = model.predict(x)
    return 'Hello ' + str(name) + ' your obesity level estimation is: ' + str(y_pred)

iface = gr.Interface(
    fn = run_my_model,
    inputs = [gr.inputs.Dropdown(choices = ['Male', 'Female']),
              gr.Slider(0,100),
             gr.Slider(0,2,.1),
             gr.Slider(39,173),
              'text',
             gr.inputs.Dropdown(choices = ['yes', 'no']),
             ],
    
    outputs = gr.outputs.Textbox(label='Obesity level: '),
    title = "Obesity Level Estimation using a SVM",
    live = True
)

iface.launch(inbrowser=True, share = True)
```
